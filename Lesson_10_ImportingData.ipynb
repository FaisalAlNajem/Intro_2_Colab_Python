## Reading CSV

```
import csv

with open(f"{sep}content{sep}Intro_2_Colab_Python{sep}Data{sep}Player.csv", newline='') as file_to_read:
  filereader = csv.reader(file_to_read, delimiter=',', quotechar='|')
  for row in filereader:
    print(len(row), row)
```

```
# We can also use Pandas 
import pandas as pd
df = pd.read_csv(f"{sep}content{sep}Intro_2_Colab_Python{sep}Data{sep}Player.csv")
df
```
